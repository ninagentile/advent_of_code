```python
def get_n_cubes_per_color(round):

    color_name = ''
    n_cubes = ''
    n_red = 0
    n_green = 0
    n_blue = 0
    for letter in round:
        if letter.isdigit():
            n_cubes += letter

        elif letter.isalpha():
            color_name += letter
            if len(color_name) >= 3:
                if 'red' in color_name:
                    n_red += int(n_cubes)
                    # print(f'found {n_cubes} red cubes')
                    color_name = ''
                    n_cubes = ''
                elif 'green' in color_name:
                    n_green += int(n_cubes)
                    # print(f'found {n_cubes} green cubes')
                    n_cubes = ''
                    color_name = ''
                elif 'blue' in color_name:
                    n_blue += int(n_cubes)
                    # print(f'found {n_cubes} blue cubes')
                    n_cubes = ''
                    color_name = ''
    return {'red': n_red, 'green': n_green, 'blue': n_blue}
```

```python
def get_power(composition):
  pow = 1
  for v in composition.values():
    pow *= v
  return pow
```

```python
def is_symbol(current_row, idx):
  try:
    symbol = current_row[idx]
    # If is not a number and is not a '.' then is a symbol
    if (not symbol.isdigit()) and (symbol != '.'):
      return True
  except Exception:
    return False

  return False

def has_adjacent_symbol(rows, n_row, el_idx):
  current_row = rows[n_row]
  up_row = None
  down_row = None
  if n_row != 0:
    up_row = rows[n_row - 1]
  if n_row != len(rows) - 1:
    down_row = rows[n_row + 1]


  # Try left and right symbols first (same row, idx+1 and idx-1)
  if is_symbol(current_row, idx=el_idx - 1) or is_symbol(current_row, idx=el_idx + 1):
    return True

  # Try up and down rows (different row, same idx)
  if is_symbol(up_row, idx=el_idx) or is_symbol(down_row, idx=el_idx):
    return True

  # Try diagonal (up row, idx - 1 or idx+1; down row idx+1, idx-1)
  if is_symbol(up_row, idx=el_idx+1) or is_symbol(up_row, idx=el_idx-1) or is_symbol(down_row, idx=el_idx+1) or is_symbol(down_row, idx=el_idx-1):
    return True


  # # Try left symbol at the start of a row
  # if is_symbol(up_row, idx=len(current_row) - 1):
  #   return True

  # # Try right symbol at the end of a row
  # if is_symbol(down_row, idx=0):
  #   return True

  return False
```

```python
def is_gear_symbol(current_row, idx):
  try:
    symbol = current_row[idx]
    if symbol == '*':
      return True
  except Exception:
    return False

  return False

def get_adjacent_gear_symbol(rows, n_row, el_idx):
  current_row = rows[n_row]
  up_row = None
  down_row = None
  up_row_n = n_row - 1
  down_row_n = n_row + 1
  if n_row != 0:
    up_row = rows[up_row_n]
  if n_row != len(rows) - 1:
    down_row = rows[down_row_n]


  # Try left and right symbols first (same row, idx+1 and idx-1)
  if is_gear_symbol(current_row, idx=el_idx - 1):
    return n_row, el_idx - 1
  if is_gear_symbol(current_row, idx=el_idx + 1):
    return n_row, el_idx + 1

  # Try up and down rows (different row, same idx)
  if is_gear_symbol(up_row, idx=el_idx):
    return up_row_n, el_idx
  if is_gear_symbol(down_row, idx=el_idx):
    return down_row_n, el_idx

  # Try diagonal (up row, idx - 1 or idx+1; down row idx+1, idx-1)
  if is_gear_symbol(up_row, idx=el_idx+1):
    return up_row_n, el_idx+1
  if is_gear_symbol(up_row, idx=el_idx-1):
    return up_row_n, el_idx-1
  if is_gear_symbol(down_row, idx=el_idx+1):
    return down_row_n, el_idx+1
  if is_gear_symbol(down_row, idx=el_idx-1):
    return down_row_n, el_idx-1

  return None
```

```python
def get_destination(mappings, seed):
    # The default destination is equal to the seed itself
    seed_destination = seed
    for mapping in mappings:
        destination = mapping[0]
        source = mapping[1]
        interval = mapping[2]

        # If the number is in the interval [source and source + interval), then
        # we need to find its destination
        if seed >= source and seed < source + interval:
            # Get the destination and return it
            seed_destination = seed_destination = destination + (seed - source)
            return seed_destination
    return seed_destination

locations =  []
for seed in seeds:
    soil = get_destination(seed_to_soil_map, seed)
    fertilizer = get_destination(soil_to_fertilizer_map, soil)
    water = get_destination(fertilizer_to_water_map, fertilizer)
    light = get_destination(water_to_light_map, water)
    temperature = get_destination(light_to_temperature_map, light)
    humidity = get_destination(temperature_to_humidity_map, temperature)
    location = get_destination(humidity_to_location_map, humidity)
    locations.append(location)
min(locations)
```

```python
from typing import List

# Map the labels of the cards to other letters so that, if put in
# lexicographic order, the first card is also the winning one
mapping_to_lexicographic_order = {
    'A': 'E', 'K': 'D', 'Q': 'C', 'J': 'B', 'T': 'A', '9': '9', '8': '8',
    '7': '7', '6': '6', '5': '5', '4': '4', '3': '3', '2': '2'
}


def is_five_of_a_kind(counts: List[int]) -> bool:
    # If all the cards have the same label, return True
    if len(counts) == 1:
        return True
    return False


def is_four_of_a_kind(counts: List[int]) -> bool:
    if counts == [1, 4]:
        return True
    return False


def is_three_of_a_kind(counts: List[int]) -> bool:
    if counts == [1, 1, 3]:
        return True
    return False


def is_full_hand(counts: List[int]) -> bool:
    if counts == [2, 3]:
        return True
    return False


def is_two_pairs(counts: List[int]) -> bool:
    if counts == [1, 2, 2]:
        return True
    return False


def is_one_pair(counts: List[int]) -> bool:
    if counts == [1, 1, 1, 2]:
        return True
    return False


def is_high_card(counts: List[int]) -> bool:
    if counts == [1, 1, 1, 1, 1]:
        return True
    return False


def get_card_counts(hand: str) -> List[int]:
    """
    Counts the occurrences of each label in hand, and returns an ordered
    list of occurrences
    """
    # If all the cards but one have the same label, return True
    unique_labels = set(hand)
    counts = []
    for unique_labels in unique_labels:
        counts.append(hand.count(unique_labels))
    return sorted(counts)


def get_mapped_hand(hand: str) -> str:
    """
    Maps a hand using mapping_to_lexicographic_order
    """
    mapped_hand = ''

    for elem in hand:
        mapped_hand += mapping_to_lexicographic_order[elem]

    return mapped_hand


def get_hands_and_bids() -> dict:
    """
    Extracts the hands and the corresponding bids from the input
    """
    hands_and_bids = {}
    rows = day_7_input.split('\n')
    for row in rows:
        hand, bid_str = row.split(' ')

        # Get the mapped version of the hand (to allow lexicographic
        # ordering)
        hand = get_mapped_hand(hand)

        hands_and_bids[hand] = int(bid_str)
    return hands_and_bids


def sort_lexicographically(hands: list) -> list:
    return sorted(hands)


def puzzle_1_solution():
    hands_and_bids = get_hands_and_bids()

    five_of_a_kinds = []
    four_of_a_kinds = []
    full_hands = []
    three_of_a_kinds = []
    two_pairs = []
    one_pairs = []
    high_cards = []

    for hand in hands_and_bids.keys():
        counts = get_card_counts(hand=hand)
        if is_five_of_a_kind(counts):
            five_of_a_kinds.append(hand)
        elif is_four_of_a_kind(counts):
            four_of_a_kinds.append(hand)
        elif is_full_hand(counts):
            full_hands.append(hand)
        elif is_three_of_a_kind(counts):
            three_of_a_kinds.append(hand)
        elif is_two_pairs(counts):
            two_pairs.append(hand)
        elif is_one_pair(counts):
            one_pairs.append(hand)
        elif is_high_card(counts):
            high_cards.append(hand)

    # Sort in lexicographic order
    five_of_a_kinds = sort_lexicographically(five_of_a_kinds)
    four_of_a_kinds = sort_lexicographically(four_of_a_kinds)
    full_hands = sort_lexicographically(full_hands)
    three_of_a_kinds = sort_lexicographically(three_of_a_kinds)
    two_pairs = sort_lexicographically(two_pairs)
    one_pairs = sort_lexicographically(one_pairs)
    high_cards = sort_lexicographically(high_cards)

    rank = 1
    total_winnings = 0
    hands_types = [
        high_cards, one_pairs, two_pairs, three_of_a_kinds,
        full_hands, four_of_a_kinds, five_of_a_kinds
    ]
    for hands_type in hands_types:
        for hand in hands_type:
            bid = hands_and_bids[hand]
            total_winnings += bid * rank
            rank += 1

    return total_winnings

puzzle_1_solution()
```

```python
from typing import List
from collections import defaultdict

# Map the labels of the cards to other letters so that, if put in
# lexicographic order, the first card is also the winning one
mapping_to_lexicographic_order = {
    'A': 'E', 'K': 'D', 'Q': 'C', 'T': 'A', '9': '9', '8': '8',
    '7': '7', '6': '6', '5': '5', '4': '4', '3': '3', '2': '2', 'J': '1'
}


def is_five_of_a_kind(counts: List[int]) -> bool:
    # If all the cards have the same label, return True
    if len(counts) == 1:
        return True
    return False


def is_four_of_a_kind(counts: List[int]) -> bool:
    if counts == [1, 4]:
        return True
    return False


def is_three_of_a_kind(counts: List[int]) -> bool:
    if counts == [1, 1, 3]:
        return True
    return False


def is_full_hand(counts: List[int]) -> bool:
    if counts == [2, 3]:
        return True
    return False


def is_two_pairs(counts: List[int]) -> bool:
    if counts == [1, 2, 2]:
        return True
    return False


def is_one_pair(counts: List[int]) -> bool:
    if counts == [1, 1, 1, 2]:
        return True
    return False


def is_high_card(counts: List[int]) -> bool:
    if counts == [1, 1, 1, 1, 1]:
        return True
    return False


def get_card_counts(hand: str) -> List[int]:
    """
    Counts the occurrences of each label in hand, and returns an ordered
    list of occurrences
    """
    # If all the cards but one have the same label, return True
    unique_labels = set(hand)
    counts = defaultdict(int)
    for unique_label in unique_labels:
        counts[unique_label] += (hand.count(unique_label))

    if mapping_to_lexicographic_order['J'] in counts.keys():
      jolly_counts = counts[mapping_to_lexicographic_order['J']]
      del counts[mapping_to_lexicographic_order['J']]

      if len(counts) == 0:
        return [jolly_counts]

      # Get the cards with the highest occurrences
      max_occurrence = max(counts.values())
      best_labels = [label for label, count in counts.items() if count == max_occurrence]
      best_label = list(reversed(sorted(best_labels)))[0]

      # Sum the count of Js to the best card
      counts[best_label] += jolly_counts

    return sorted(list(counts.values()))


def get_mapped_hand(hand: str) -> str:
    """
    Maps a hand using mapping_to_lexicographic_order
    """
    mapped_hand = ''

    for elem in hand:
        mapped_hand += mapping_to_lexicographic_order[elem]

    return mapped_hand


def get_hands_and_bids() -> dict:
    """
    Extracts the hands and the corresponding bids from the input
    """
    hands_and_bids = {}
    rows = day_7_input.split('\n')
    for row in rows:
        hand, bid_str = row.split(' ')

        # Get the mapped version of the hand (to allow lexicographic
        # ordering)
        hand = get_mapped_hand(hand)

        hands_and_bids[hand] = int(bid_str)
    return hands_and_bids


def sort_lexicographically(hands: list) -> list:
    return sorted(hands)


def puzzle_2_solution():
    hands_and_bids = get_hands_and_bids()

    five_of_a_kinds = []
    four_of_a_kinds = []
    full_hands = []
    three_of_a_kinds = []
    two_pairs = []
    one_pairs = []
    high_cards = []

    for hand in hands_and_bids.keys():
        counts = get_card_counts(hand=hand)
        if is_five_of_a_kind(counts):
            five_of_a_kinds.append(hand)
        elif is_four_of_a_kind(counts):
            four_of_a_kinds.append(hand)
        elif is_full_hand(counts):
            full_hands.append(hand)
        elif is_three_of_a_kind(counts):
            three_of_a_kinds.append(hand)
        elif is_two_pairs(counts):
            two_pairs.append(hand)
        elif is_one_pair(counts):
            one_pairs.append(hand)
        elif is_high_card(counts):
            high_cards.append(hand)

    # Sort in lexicographic order
    five_of_a_kinds = sort_lexicographically(five_of_a_kinds)
    four_of_a_kinds = sort_lexicographically(four_of_a_kinds)
    full_hands = sort_lexicographically(full_hands)
    three_of_a_kinds = sort_lexicographically(three_of_a_kinds)
    two_pairs = sort_lexicographically(two_pairs)
    one_pairs = sort_lexicographically(one_pairs)
    high_cards = sort_lexicographically(high_cards)

    rank = 1
    total_winnings = 0
    hands_types = [
        high_cards, one_pairs, two_pairs, three_of_a_kinds,
        full_hands, four_of_a_kinds, five_of_a_kinds
    ]
    for hands_type in hands_types:
        for hand in hands_type:
            bid = hands_and_bids[hand]
            total_winnings += bid * rank
            rank += 1

    return total_winnings

puzzle_2_solution()
```

```python
start = 'AAA'
end = 'ZZZ'

def get_steps(start, end, instructions, network):
  current_node = start
  steps = 0
  instructions_ended = False
  instruction_idx = 0

  while True:
    current_instruction = instructions[instruction_idx]

    # Go left or right
    if current_instruction == 'L':
      current_node = network[current_node][0]
    else:
      current_node = network[current_node][1]

    steps += 1

    # Check if ended
    if (current_node == end) and (instruction_idx == len(instructions) - 1):
      break
    # If you ended the instructions without finding the end node, restart
    if instruction_idx == len(instructions) - 1:
      instruction_idx = 0
    else:
      instruction_idx += 1
  return steps
get_steps(start, end, instructions, network)
```
