```python
def get_n_cubes_per_color(round):

    color_name = ''
    n_cubes = ''
    n_red = 0
    n_green = 0
    n_blue = 0
    for letter in round:
        if letter.isdigit():
            n_cubes += letter

        elif letter.isalpha():
            color_name += letter
            if len(color_name) >= 3:
                if 'red' in color_name:
                    n_red += int(n_cubes)
                    # print(f'found {n_cubes} red cubes')
                    color_name = ''
                    n_cubes = ''
                elif 'green' in color_name:
                    n_green += int(n_cubes)
                    # print(f'found {n_cubes} green cubes')
                    n_cubes = ''
                    color_name = ''
                elif 'blue' in color_name:
                    n_blue += int(n_cubes)
                    # print(f'found {n_cubes} blue cubes')
                    n_cubes = ''
                    color_name = ''
    return {'red': n_red, 'green': n_green, 'blue': n_blue}
```

```python
def get_power(composition):
  pow = 1
  for v in composition.values():
    pow *= v
  return pow
```

```python
def is_symbol(current_row, idx):
  try:
    symbol = current_row[idx]
    # If is not a number and is not a '.' then is a symbol
    if (not symbol.isdigit()) and (symbol != '.'):
      return True
  except Exception:
    return False

  return False

def has_adjacent_symbol(rows, n_row, el_idx):
  current_row = rows[n_row]
  up_row = None
  down_row = None
  if n_row != 0:
    up_row = rows[n_row - 1]
  if n_row != len(rows) - 1:
    down_row = rows[n_row + 1]


  # Try left and right symbols first (same row, idx+1 and idx-1)
  if is_symbol(current_row, idx=el_idx - 1) or is_symbol(current_row, idx=el_idx + 1):
    return True

  # Try up and down rows (different row, same idx)
  if is_symbol(up_row, idx=el_idx) or is_symbol(down_row, idx=el_idx):
    return True

  # Try diagonal (up row, idx - 1 or idx+1; down row idx+1, idx-1)
  if is_symbol(up_row, idx=el_idx+1) or is_symbol(up_row, idx=el_idx-1) or is_symbol(down_row, idx=el_idx+1) or is_symbol(down_row, idx=el_idx-1):
    return True


  # # Try left symbol at the start of a row
  # if is_symbol(up_row, idx=len(current_row) - 1):
  #   return True

  # # Try right symbol at the end of a row
  # if is_symbol(down_row, idx=0):
  #   return True

  return False
```

```python
def is_gear_symbol(current_row, idx):
  try:
    symbol = current_row[idx]
    if symbol == '*':
      return True
  except Exception:
    return False

  return False

def get_adjacent_gear_symbol(rows, n_row, el_idx):
  current_row = rows[n_row]
  up_row = None
  down_row = None
  up_row_n = n_row - 1
  down_row_n = n_row + 1
  if n_row != 0:
    up_row = rows[up_row_n]
  if n_row != len(rows) - 1:
    down_row = rows[down_row_n]


  # Try left and right symbols first (same row, idx+1 and idx-1)
  if is_gear_symbol(current_row, idx=el_idx - 1):
    return n_row, el_idx - 1
  if is_gear_symbol(current_row, idx=el_idx + 1):
    return n_row, el_idx + 1

  # Try up and down rows (different row, same idx)
  if is_gear_symbol(up_row, idx=el_idx):
    return up_row_n, el_idx
  if is_gear_symbol(down_row, idx=el_idx):
    return down_row_n, el_idx

  # Try diagonal (up row, idx - 1 or idx+1; down row idx+1, idx-1)
  if is_gear_symbol(up_row, idx=el_idx+1):
    return up_row_n, el_idx+1
  if is_gear_symbol(up_row, idx=el_idx-1):
    return up_row_n, el_idx-1
  if is_gear_symbol(down_row, idx=el_idx+1):
    return down_row_n, el_idx+1
  if is_gear_symbol(down_row, idx=el_idx-1):
    return down_row_n, el_idx-1

  return None
```

```python
def get_destination(mappings, seed):
    # The default destination is equal to the seed itself
    seed_destination = seed
    for mapping in mappings:
        destination = mapping[0]
        source = mapping[1]
        interval = mapping[2]

        # If the number is in the interval [source and source + interval), then
        # we need to find its destination
        if seed >= source and seed < source + interval:
            # Get the destination and return it
            seed_destination = seed_destination = destination + (seed - source)
            return seed_destination
    return seed_destination

locations =  []
for seed in seeds:
    soil = get_destination(seed_to_soil_map, seed)
    fertilizer = get_destination(soil_to_fertilizer_map, soil)
    water = get_destination(fertilizer_to_water_map, fertilizer)
    light = get_destination(water_to_light_map, water)
    temperature = get_destination(light_to_temperature_map, light)
    humidity = get_destination(temperature_to_humidity_map, temperature)
    location = get_destination(humidity_to_location_map, humidity)
    locations.append(location)
min(locations)
```
